```python
# DataFrame to store all imported data
import os
if not os.path.isfile('data.csv'):
    data = open('data.csv', mode='w')

files = ['combined_data_1.txt',
#        'combined_data_2.txt',
#          '../input/netflix-prize-data/combined_data_3.txt',  # read in half of data only for speedup
#          '../input/netflix-prize-data/combined_data_4.txt'
        ]

# Remove the line with movie_id: and add a new column of movie_id
# Combine all data files into a csv file
for file in files:
  print("Opening file: {}".format(file))
  with open(file) as f:
    for line in f:
        line = line.strip()
        if line.endswith(':'):
            movie_id = line.replace(':', '')
        else:
            data.write(movie_id + ',' + line)
            data.write('\n')
data.close()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.sparse import csr_matrix
%matplotlib inline
```

```python
df  =pd.read_csv('data.csv', names = ['movie_id','cust_id', 'rating','year'])
movies = pd.read_csv('movie_titles.csv',encoding = "ISO-8859-1", header = None, error_bad_lines=False, names = ['movie_id','year','title'])
```

```python

def create_matrix(df):
    N = df['cust_id'].nunique()
    M = df['movie_id'].nunique()
    # map ids to indices
    user_mapper = dict(zip(np.unique(df['cust_id']), list(range(N))))
    movie_mapper = dict(zip(np.unique(df['movie_id']), list(range(M))))

    # map ids to indices
    user_inv_mapper = dict(zip( list(range(N)),np.unique(df['cust_id'])))
    movie_inv_mapper = dict(zip( list(range(M)),np.unique(df['movie_id'])))

    user_index = [user_mapper[i] for i in df['cust_id']]
    movie_index = [movie_mapper[i] for i in df['movie_id']]
    
    X = csr_matrix((df['rating'], (movie_index, user_index)), shape=(M, N))
    return X, user_mapper, movie_mapper, user_inv_mapper, movie_inv_mapper, movie_index, user_index
```

```python
X, user_mapper, movie_mapper, user_inv_mapper, movie_inv_mapper, user_index, movie_index = create_matrix(df)
```

```python
from sklearn.neighbors import NearestNeighbors

# find similar movies using KNN
def find_similar_movies(movie_id, X, k):
    neighbour_ids = []

    movie_ind = movie_mapper[movie_id]
    movie_vec = X[movie_ind]

    k+=1

    KNN = NearestNeighbors(n_neighbors= k, algorithm='brute', metric= 'cosine')
    KNN.fit(X)
    movie_vec = movie_vec.reshape(1,-1)
    neighbour = KNN.kneighbors(movie_vec, return_distance=False)

    for i in range(0,k):
        n = neighbour.item(i)
        neighbour_ids.append(movie_inv_mapper[n])

    neighbour_ids.pop(0)
    return neighbour_ids
```

```python
movie_titles = dict(zip(movies['movie_id'], movies['title']))
movie_id = 3
```

```python
similar_ids = find_similar_movies(movie_id, X, k=10)
movie_title = movie_titles[movie_id]
```

```python
print(f'Since you watched {movie_title}')
for i in similar_ids:
    print(movie_titles[i])
```
